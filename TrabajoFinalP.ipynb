## 2. Carga de datos

El primer paso del proyecto, es la carga de nuestros datasets en Python, para esto utilizaremos `Pandas`, cabe destacar que hay que seleccionar el encoding correspondiente y tambien el separador que ha sido utilizado.

```python
# Carga de los 2 csv
import pandas as pd
bici_2016 = pd.read_csv ('Recursos/AccidentesBicicletas_2016.csv', encoding = "iso-8859-1",sep =';', index_col=False)
bici_2017 = pd.read_csv ('Recursos/AccidentesBicicletas_2017.csv', encoding = "iso-8859-1", sep =';', index_col=False)
```

### 2.1 Concatenamos los datos

Juntamos ambos archivos y los convertimos en una unica tabla, creando el indice correspondiente para la misma.

```python
# Concatenamos los 2csv e indexar
Final = pd.concat([bici_2016, bici_2017], axis = 0, sort = False, ignore_index = True)
Final
```

## 3. Tratamiendo de los datos.

A partir de este punto, ya tenemos nuestra tabla creada, el siguiente paso es prepararla para poder realizar el futuro analisis.

### 3.1 Cambiar el nombre de las columnas

Quitamos los espacios y simbolos extraños de los nombres de las columnas, renombrandolos todos ellos.

```python
# Cambiamos los nombres de las columnas
Final.columns = ['Fecha','R_Horario','Dia_Semana','Distrito','Lugar','Numero','Numero_Parte','FA_Granizo','FA_Hielo','FA_Lluvia','FA_Niebla','FA_Seco','FA_Nieve','SV_Mojado','SV_Aceite','SV_Barro','SV_Grava','SV_Hielo','SV_Seco_limpio','Victimas','Tipo','Vehiculo','Persona','Sexo','Lesividad','Edad']
Final.columns
Final
```

## 3.2 Visualización del tipo de datos que tenemos


```python
Final.dtypes # Nos dice de qué tipo de dato es cada variable
```

```python
Final['Dia'], Final['Mes'] = Final['Fecha'].str.split('/', 1).str
Final['Mes'], Final['Ano'] = Final['Mes'].str.split('/', 1).str
Final
```

## 3.3 Cambiamos el formato de la fecha

La fecha nos aparece como un `Object`, cuando lo correcto seria que fuese un `Datetime`.

```python
# Cambiamos 'fecha' a formato datetime
from datetime import date, datetime
Final1 = Final.copy()
Final1["Fecha"] = pd.to_datetime(Final1["Fecha"], format = "%d/%m/%Y")
Final1.dtypes
```

```python
## 3.4 Eliminamos las filas con nulos
```

```python
Final1 = Final1.dropna() # Eliminamos las filas que contengan valores nulos
len(Final1)
```

```python
# Cambiamos el formato de R_Horario

import pandas as pd

morning = Final1.R_Horario.apply(lambda i: i.replace("DE ", ""))
apertura = []
cierre = []
for i in morning:
    tmp = i.split("A")
    apertura.append(tmp[0].strip())
#     cierre.append(tmp[1].strip())


Final1['R_Horario']= apertura

Final1


# si necesitásemos el horario de cierre, se encuentra en la variable "cierre"
```

## 3.5 Procedemos a cambiar el formato de edad

```python
# Cambiamos el formato de Edad (Eliminamos el rango de +74 años)

import pandas as pd

Final2 = Final1[(Final1.Edad != 'DESCONOCIDA') & (Final1.Edad != 'DE MAS DE 74')]

Final2
```

```python

filtrado = Final2.Edad.apply(lambda x: x[3:])

columna_sin_vacios = Final2['Edad'][Final2['Edad'].notnull()] # Quitamos los elementos nulos

columna_filtrada = columna_sin_vacios.apply(lambda x: x[3:]) # Eliminamos los 3 primeros elementos de la cadena, es decir, 'DE '

columna_filtrada = columna_filtrada.apply(lambda x: x.split(' ')) # Partimos la cadena en 4 separando por espacio ' '

# # Ahora tenemos una lista de arrays con cuatro elementos cada uno

edad_1 = [i[0].replace(' ','') for i in columna_filtrada] # Nos quedamos con la edad_1

edad_2 = [i[2].replace(' ','') for i in columna_filtrada] # Nos quedamos con la edad_2

Final2.Edad = edad_2

Final2
```

## 3.6 Cambiamos la edad a Float

Por ultimo, cambiamos de nuevo el formato de la columna "Edad", para poder trabajar con ella en el analisis.

```python
Final2.Edad = Final2.Edad.apply(lambda x: float (x))
Final2.dtypes
```

# 4.Análisis exploratorio de los datos

El primer paso a la hora de realizar el analisis es la carga de las librerias correspondientes, en este caso utilizaremos `Seaborn`, `Matploit`. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['figure.figsize']= 10, 9
sns.set_context('talk')

Grafico1 = sns.countplot(x='Sexo', hue='Tipo', data=Final2)

# En el siguiente grafico analizamos el numero de persona por sexo, que se agrupo por tipo de accidente
```

## 4.1 Lesividad por victimas

Observamos diferentes grados de lesividad que se han dado en los accidentes.

    - IL : Impacto leve
    - HL : Heridas leves
    - HG : Heridas graves
    - MT : Muerto
 Y por ultimo un porcentaje de accidentes donde la lesividad no ha sido registrada, los cuales no se tendran en cuenta.

```python
g = sns.FacetGrid(Final2, col="Lesividad", sharey=False) # Representamos los quilates agrupados por cut en diferentes gráficas
g.map(plt.hist, "Victimas")
```

Podemos observar que las heridas leves son las mas habituales en este tipo de accidentes, a su ves observamos que en muy pocas ocasiones se producen accidentes con varias victimas, ya que en estos casos, el conductor de la bicicleta suele ser el mayor perjudicado.

## 4.2 Numero de victimas por dia de la semana

En el siguiente grafico, se muestran los dias de las semana y cuantas victimas se producen en ellos:

```python
rcParams['figure.figsize']= 15, 9
Grafico3 = sns.violinplot(x='Dia_Semana', y='Victimas', data=Final2)
```

Podemos observar con claridad que el dia Jueves, es el dia en el que se producen mas accidentes de la semana.
A su vez, se ve que en el fin de semana, en especial en el domingo es cuando se producen los accidentes mas graves, ya que el numero de victimas es mayor, produciendose accidentes con hasta 3 victimas implicadas.

## 4.3 Suelo mojado frente a suelo seco

A continuacion, mostramos una comparativa haciendo uso de un espectro de los accidentes que se producen con el suelo seco y con el suelo mojado.

```python
Grafico4 = sns.lmplot(x = 'Victimas', y ='Edad', hue ='SV_Mojado', data=Final2)
```

Atendiendo al espectro generado, podemos observar que es muchisimo mayor la amplitud cuando llueve, lo que significa que los accidentes de producen en un rango mayor de edad, mientras que cuando el suelo no esta mojado, afecta a un rango menor de edades, lo que nos puede dar a entender que, las personas mayores tienen una mayor propensión a sufrir accidentes de bicicleta cuando el pavimento se encuenta mojado.

# 4.4 Accidentes por horas del dia

## 4.5 Tabla comparativa de victimas totales

```python
total = Final2.groupby('Ano').Victimas.apply(lambda x: x.sum()).reset_index()
Comparativa_Victimas = pd.DataFrame(total)
Comparativa_Victimas = Comparativa_Victimas.sort_values('Victimas', ascending=False)
Comparativa_Victimas
```

## 4.6 Tabla y grafico de victimas por distritos

```python
total1 = Final2.groupby('Distrito').Victimas.apply(lambda x: x.sum()).reset_index()
Comparativa_Distrito = pd.DataFrame(total1)
Comparativa_Distrito = Comparativa_Distrito.sort_values('Victimas', ascending=False)
Comparativa_Distrito
```

```python
out = sns.barplot(x='Distrito', y='Victimas', data=Comparativa_Distrito)
out= plt.xticks(rotation = 90)
```

Observamos que el distrito Centro es el que mas accidentes aglutina con 257 accidentes, el que menos accidentes ocasiona es Moratalaz con 18 accidentes.

## 4.7 Indice de Lesividad

```python
total2 = Final2.groupby('Lesividad').Victimas.apply(lambda x: x.sum())
Comparativa_Lesividad = pd.DataFrame(total2)
Comparativa_Lesividad = Comparativa_Lesividad.sort_values('Victimas', ascending=False).reset_index()
Comparativa_Lesividad
```

```python
n = sns.barplot(x='Lesividad', y='Victimas', data=Comparativa_Lesividad)
```

Como anteriormente hemos dicho, aqui demostramos con mas precision que la mayoria de lesiones que se producen en los accidentes son heridas leves.
